```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#import file dari csv
df_customers = pd.read_csv(r'D:\Tableu\visualisasi_py\Customers.csv',index_col=0)
df_customers.head()
```

```python
#cek column 
df_customers.info()
df_customers.columns
```

```python
#cek column gender.jumlah
gender = df_customers['Gender'].value_counts()
gender

```

```python
plot = gender.plot.pie(subplots=False, figsize=(3, 3),autopct='%1.0f%%')
gender
```

```python
profesi = df_customers['Profession'].value_counts()
profesi
```

```python
bar_profesi = profesi.plot.bar(x='profesi',y='val', figsize = (7,4))
profesi
```

```python
scatter = df_customers.plot.scatter(x = 'Age',
                                    y = 'Annual Income ($)',
                                    c='Work Experience',
                                    colormap='viridis',
                                    figsize=(15, 5))
```

```python
avg_prof = df_customers.loc[:,['Profession','Annual Income ($)']]
avg_prof.groupby(by='Profession').mean().round(2).sort_values(by='Annual Income ($)', ascending=False)
```

```python
jumlah_prof = df_customers.loc[:,['Profession','Annual Income ($)']]
jumlah_prof1=jumlah_prof.groupby(by='Profession').sum('Annual Income ($)')
jumlah_prof1
```
